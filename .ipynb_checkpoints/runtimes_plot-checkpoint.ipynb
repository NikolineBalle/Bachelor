```python
import numpy as np
from numpy.linalg import solve
import matplotlib.pyplot as plt
from scipy.linalg import qr
from scipy import optimize
from scipy.special import erfinv
from scipy.optimize import minimize_scalar
import egrss 
import time
import lambda_fun 
import os



def plot_deg(y, col, xlabel, ylabel, savename):
    fig = plt.figure()
    ax = fig.add_axes([0, 0,  1, 1])
    bp = ax.boxplot(y)
    
    ax = plt.gca()
    ax.set_facecolor('whitesmoke')
    plt.xlabel(xlabel, fontsize = 22)
    plt.ylabel(ylabel, fontsize = 22)
    
    plt.yticks(fontsize = 22)
    plt.xticks(fontsize = 22)
    plt.tick_params(left=True, bottom = True, right = False, top = False)
    ax = plt.gca()
    ax.set_facecolor('whitesmoke')
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
    plt.show
    
    my_path = os.getcwd()
    fig.savefig(my_path + savename, bbox_inches='tight')
    
  
plot_deg(GML_deg, 'cornflowerblue' ,xlabel = 's', ylabel = 'time [s]',savename = "/Plots/GML_deg.pdf")

plot_deg(GCV_deg,'cornflowerblue', xlabel = 's', ylabel = 'time [s]', savename = "/Plots/GCV_deg.pdf")


def plot_N(x,y, col, xlabel, ylabel, savename):
    fig = plt.figure()
    x1 = np.linspace(np.min(x), np.max(x), 100)
    y0mean = np.mean(y[0,:])
    y1mean = np.mean(y[1,:])
    plt.loglog(x1,y0mean + (y1mean-y0mean)/(x[1]-x[0])*x1, '-' , color= 'hotpink', label = '$\mathcal{O}(N)$')
    plt.loglog(x1, y0mean+  x1**2 , '-' , color= 'mediumpurple', label = '$\mathcal{O}(N^2)$')
    plt.legend(loc = 'best')
    for i in range(y.shape[0]):
        plt.loglog(x,y[:,i], '.' ,color = col, markersize = 4.5, alpha = 0.4)

    ax = plt.gca()
    ax.set_facecolor('whitesmoke')
    plt.xlabel(xlabel, fontsize = 15)
    plt.ylabel(ylabel, fontsize = 15)
    
    plt.yticks(fontsize = 15)
    plt.tick_params(left=True, bottom = True, right = False, top = False)
    ax = plt.gca()
    ax.set_facecolor('whitesmoke')
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
    plt.show
    
    my_path = os.getcwd()
    fig.savefig(my_path + savename, bbox_inches='tight')
    
    


plot_N(numpoints, GML_N, 'navy', xlabel = 'log(N)', ylabel = 'log(time) [s]', savename = "/Plots/GML_N.pdf")

plot_N(numpoints, GCV_N, 'navy', xlabel = 'log(N)', ylabel = 'log(time) [s]', savename = "/Plots/GCV_N.pdf")
```
